```python
import pandas as pd
from pandas import DataFrame, Series
```

```python
def parse_row(row):
    return {
        'station': row[0:11],
        'month': row[12:14],
        'day': row[15:17],
        '00': float(row[18:23])/10,
        '01': float(row[25:30])/10,
        '02': float(row[32:37])/10,
        '03': float(row[39:44])/10,
        '04': float(row[46:51])/10,
        '05': float(row[53:58])/10,
        '06': float(row[60:65])/10,
        '07': float(row[67:72])/10,
        '08': float(row[74:79])/10,
        '09': float(row[81:86])/10,
        '10': float(row[88:93])/10,
        '11': float(row[95:100])/10,
        '12': float(row[102:107])/10,
        '13': float(row[109:114])/10,
        '14': float(row[116:121])/10,
        '15': float(row[123:128])/10,
        '16': float(row[130:135])/10,
        '17': float(row[137:142])/10,
        '18': float(row[144:149])/10,
        '19': float(row[151:156])/10,
        '20': float(row[158:163])/10,
        '21': float(row[165:170])/10,
        '22': float(row[172:177])/10,
        '23': float(row[179:184])/10,
    }
```

```python
dewp = pd.read_csv('/Users/tjl/Downloads/hly-dewp-normal.txt', header=None, delimiter='\t')
temp = pd.read_csv('/Users/tjl/Downloads/hly-temp-normal.txt', header=None, delimiter='\t')
wind = pd.read_csv('/Users/tjl/Downloads/hly-wind-avgspd.txt', header=None, delimiter='\t')
df_dewp = DataFrame(dewp[0].apply(parse_row).to_dict()).T
df_temp = DataFrame(temp[0].apply(parse_row).to_dict()).T
df_wind = DataFrame(wind[0].apply(parse_row).to_dict()).T
```

```python
df_dewp.shape, df_temp.shape, df_wind.shape
```

```python
df = pd.merge(df_dewp, df_temp, on=['station', 'month', 'day'], suffixes=('_d', '_t'))
df.head(5)
```

```python
df_wind.columns = [x+'_w' for x in df_wind.columns if x not in ['day', 'month', 'station']] + ['day', 'month', 'station']
df_wind.columns
```

```python
df = pd.merge(df, df_wind, on=['station', 'month', 'day'])
df.head(5)
```

```python
def calc_mean(row):
    nums = []
    if abs(row['00']) < 500: nums.append(row['00'])
    if abs(row['01']) < 500: nums.append(row['01'])
    if abs(row['02']) < 500: nums.append(row['02'])
    if abs(row['03']) < 500: nums.append(row['03'])
    if abs(row['04']) < 500: nums.append(row['04'])
    if abs(row['05']) < 500: nums.append(row['05'])
    if abs(row['06']) < 500: nums.append(row['06'])
    if abs(row['07']) < 500: nums.append(row['07'])
    if abs(row['08']) < 500: nums.append(row['08'])
    if abs(row['09']) < 500: nums.append(row['09'])
    if abs(row['10']) < 500: nums.append(row['10'])
    if abs(row['11']) < 500: nums.append(row['11'])
    if abs(row['12']) < 500: nums.append(row['12'])
    if abs(row['13']) < 500: nums.append(row['13'])
    if abs(row['14']) < 500: nums.append(row['14'])
    if abs(row['15']) < 500: nums.append(row['15'])
    if abs(row['16']) < 500: nums.append(row['16'])
    if abs(row['17']) < 500: nums.append(row['17'])
    if abs(row['18']) < 500: nums.append(row['18'])
    if abs(row['19']) < 500: nums.append(row['19'])
    if abs(row['20']) < 500: nums.append(row['20'])
    if abs(row['21']) < 500: nums.append(row['21'])
    if abs(row['22']) < 500: nums.append(row['22'])
    if abs(row['23']) < 500: nums.append(row['23'])
    return sum(nums) / len(nums) if len(nums) > 0 else 0

df['00'] = df['00_d'] + df['00_t'] - df['00_w']
df['01'] = df['01_d'] + df['01_t'] - df['01_w']
df['02'] = df['02_d'] + df['02_t'] - df['02_w']
df['03'] = df['03_d'] + df['03_t'] - df['03_w']
df['04'] = df['04_d'] + df['04_t'] - df['04_w']
df['05'] = df['05_d'] + df['05_t'] - df['05_w']
df['06'] = df['06_d'] + df['06_t'] - df['06_w']
df['07'] = df['07_d'] + df['07_t'] - df['07_w']
df['08'] = df['08_d'] + df['08_t'] - df['08_w']
df['09'] = df['09_d'] + df['09_t'] - df['09_w']
df['10'] = df['10_d'] + df['10_t'] - df['10_w']
df['11'] = df['11_d'] + df['11_t'] - df['11_w']
df['12'] = df['12_d'] + df['12_t'] - df['12_w']
df['13'] = df['13_d'] + df['13_t'] - df['13_w']
df['14'] = df['14_d'] + df['14_t'] - df['14_w']
df['15'] = df['15_d'] + df['15_t'] - df['15_w']
df['16'] = df['16_d'] + df['16_t'] - df['16_w']
df['17'] = df['17_d'] + df['17_t'] - df['17_w']
df['18'] = df['18_d'] + df['18_t'] - df['18_w']
df['19'] = df['19_d'] + df['19_t'] - df['19_w']
df['20'] = df['20_d'] + df['20_t'] - df['20_w']
df['21'] = df['21_d'] + df['21_t'] - df['21_w']
df['22'] = df['22_d'] + df['22_t'] - df['22_w']
df['23'] = df['23_d'] + df['23_t'] - df['23_w']
df['mean'] = df.apply(calc_mean, axis=1)
```
